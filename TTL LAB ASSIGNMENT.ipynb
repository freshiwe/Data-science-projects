```python
import scipy.integrate as spi

def f(x):
    return x**2

result, _ = spi.quad(f, 0, 1)
print(f"Integral of x^2 from 0 to 1: {result:.4f}")
```

```python
import numpy as np
from scipy.misc import derivative

def g(x):
    return np.sin(x)

derivative_at_pi_over_4 = derivative(g, np.pi/4, dx=1e-6)
print(f"Derivative of sin(x) at x = pi/4: {derivative_at_pi_over_4:.4f}")
```

```python
from scipy.optimize import minimize

def h(x):
    return x**2 + 3*x + 5

result = minimize(h, x0=0)
print(f"Minimum value of h(x): {result.fun:.4f} at x = {result.x[0]:.4f}")
```

```python
import numpy as np
from scipy.linalg import solve

# Example system of equations: 2x + y = 5, 3x - 2y = 8
A = np.array([[2, 1], [3, -2]])
b = np.array([5, 8])

x = solve(A, b)
print(f"Solutions (x, y): ({x[0]:.4f}, {x[1]:.4f})")
```

```python
import cv2

# Load an image from file
img = cv2.imread(r"C:\Users\KIIT\Downloads\business_2_18.png")

# Display the image in a window
cv2.imshow('Loaded Image', img)

# Wait for a key press to close the window
cv2.waitKey(0)

# Close all OpenCV windows
cv2.destroyAllWindows()
```

```python
# Resize the image to a new width and height
new_width, new_height = 300, 200
resized_img = cv2.resize(img, (new_width, new_height))

# Display the resized image
cv2.imshow('Resized Image', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotate the image by 45 degrees clockwise
angle = 45
rows, cols = resized_img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
rotated_img = cv2.warpAffine(resized_img, rotation_matrix, (cols, rows))

# Display the rotated image
cv2.imshow('Rotated Image', rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Flip the image vertically
flipped_img = cv2.flip(rotated_img, 0)

# Display the flipped image
cv2.imshow('Flipped Image', flipped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur
blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

# Convert BGR to RGB for display
blurred_img_rgb = cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB)

# Display the blurred image
plt.figure(figsize=(6, 6))
plt.imshow(blurred_img_rgb)
plt.title('Gaussian Blur')
plt.axis('off')
plt.show()
```

```python
# Apply median blur
median_blurred_img = cv2.medianBlur(img, 5)

# Convert BGR to RGB for display
median_blurred_img_rgb = cv2.cvtColor(median_blurred_img, cv2.COLOR_BGR2RGB)

# Display the median-blurred image
plt.figure(figsize=(6, 6))
plt.imshow(median_blurred_img_rgb)
plt.title('Median Blur')
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale using pixel manipulation (average method)
(row, col) = img.shape[0:2]
for i in range(row):
    for j in range(col):
        img[i, j] = sum(img[i, j]) // 3

# Display the grayscale image
cv2.imshow('Grayscale Image (Pixel Manipulation)', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Apply Gaussian blur to reduce noise
blurred_image = cv2.GaussianBlur(img, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred_image, 30, 100)  # Adjust thresholds as needed

# Display the original image and its edges
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.axis('off')
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Edges (Canny)'), plt.axis('off')
plt.show()
```

```python
# Find contours in the edges image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original RGB image
contour_image = img.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)  # Green color, thickness 2

# Display the image with contours
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Contours Detected')
plt.axis('off')
plt.show()
```
